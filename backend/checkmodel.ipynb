```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
from PIL import Image

# Load model
model = load_model("../models/vgg16_fintuned.h5", compile=False)

# Class names
class_names = ["Glioma", "Meningioma", "No Tumor", "Pituitary"]

# Prediction function
def predict_mri(file_path):
    # Load and preprocess image
    img = Image.open(file_path).convert("RGB")
    img = img.resize((224, 224))  # adjust to match model input size
    x = np.array(img) / 255.0
    x = np.expand_dims(x, axis=0)
    
    # Predict
    preds = model.predict(x)
    pred_index = np.argmax(preds[0])
    confidence_score = float(preds[0][pred_index] * 100)
    
    tumor_type = class_names[pred_index]
    tumor_detected = tumor_type != "No Tumor"
    tumor_location = "Frontal Lobe" if tumor_detected else None
    analysis_notes = (
        f"{tumor_type} detected. Please consult a medical professional."
        if tumor_detected else
        "No tumor detected. Brain tissue appears normal."
    )
    
    return {
        "tumor_detected": tumor_detected,
        "confidence_score": confidence_score,
        "tumor_type": tumor_type if tumor_detected else None,
        "tumor_location": tumor_location,
        "analysis_notes": analysis_notes
    }

# Test with a sample image
file_path = "../backend/img.jpg"  # replace with your MRI image file path
result = predict_mri(file_path)
print(result)
```
